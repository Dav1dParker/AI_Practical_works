# 4.3.5 Задание

```python
import math
def summ(a,b):
    return a + b
def subtraction(a,b):
    return a - b
def multiplication(a,b):
    return a * b
def division(a,b):
    return a/b
def func1(a,b):
    return math.exp(a+b)
def func2(a,b):
    return math.sin(a + b)
def func3(a,b):
    return math.cos(a + b)
def func4(a,b):
    return a ** b

a, s, b = input().split()

a = float(a)
b = float(b)
if s == '+':
    ans = summ(a,b)
if s == '-':
    ans = subtraction(a,b)
if s == '*':
    ans = multiplication(a,b)
if s == '/':
    ans = division(a,b)
if s == "f1":
    ans = func1(a,b)
if s == "f2":
    ans = func2(a,b)
if s == "f3":
    ans = func3(a,b)
if s == "f4":
    ans = func4(a,b)
print(a,s,b,"=",ans)
```
